### 【태그내에서 여러 속성값이 있을때는 】
- soup.find_all('태그명', attrs={'태그속성타이틀':'속성'})

```python
from bs4 import BeautifulSoup
ttt="""
 <body>
 <p id='a' align='rigth'> 오른쪽 정렬1 </p>
 <p id='b' align='rigth'> 오른쪽 정렬2 </p>
 <p id='a' align='center'> 가운데 정렬1 </p>
 <p id='b' align='center'> 가운데 정렬2 </p>
 <p id='b' align='center'> 가운데 정렬3 </p>
 <body>


"""

# HTML 태그와 태그에 있는 속성:속성값을 활용해서 필요한 데이터를 추출하는 방법

soup = BeautifulSoup(ttt, "html.parser")
soup
re= soup.find_all('p', attrs = {'align': 'center'})
print('\n',re)

re1= soup.find_all('p', attrs = {'align': 'center', 'id':'b'})
print('\n',re)
```

```python
print(re[0].get_text())
print(re[1].get_text())
print(re[2].get_text())
```

```python
# fild_all 자료의 리스트 구조는 for를 이용하여서 출력해야함

for i in re:
    print(i.get_text())
```

```python
# fild_all 자료의 리스트 구조는 for를 이용하여서 출력해야함
# for구문은 한개로 작성할수 있음.

[i.get_text() for i in re]
```
