```python
import requests
import json
from bs4 import BeautifulSoup
import pandas as pd

# 네이버 쇼핑 API 키와 엔드포인트 설정
naver_api_key = 'zU8Zvu5B5kb0xxz6zwzA'
endpoint = 'https://openapi.naver.com/v1/search/shop.json'

# 검색 조건 설정
params = {
    'query': 'LG스마트TV',
    'display': 100,  # 가져올 아이템 개수
    'start': 1,    # 시작 위치
    'sort': 'sim',  # 정렬 방식
}

# API 요청 보내기
headers = {'X-Naver-Client-Id': 'zU8Zvu5B5kb0xxz6zwzA', 'X-Naver-Client-Secret': 'Fze5ceyruX'}
response = requests.get(endpoint, params=params, headers=headers)

# 응답 처리
if response.status_code == 200:
    result = json.loads(response.text)
    items = result.get('items', [])

    # 데이터를 저장할 리스트
    data = []

    for item in items:
        # BeautifulSoup을 사용하여 HTML 태그 제거
        product_name = BeautifulSoup(item.get('title', ''), 'html.parser').get_text()
        # 상품명을 소문자로 변환하고 공백 제거
        product_name_normalized = product_name.lower().replace(' ', '')
        lowest_price = item.get('lprice', '')
        mall_name = item.get('mallName', '')
        mall_link = item.get('link', '')

        # 각 아이템을 데이터 리스트에 추가
        data.append({
            "정규화된 상품명": product_name_normalized,
            "원래 상품명": product_name,
            "최저가": lowest_price,
            "쇼핑몰명": mall_name,
            "링크": mall_link
        })

    # 판다스 데이터프레임으로 변환
    df = pd.DataFrame(data)

    # 판다스 출력 옵션 설정
    pd.set_option('display.max_rows', None)

    print(df)
else:
    print(f"Error: {response.status_code}")
```
